#### création d'une collection Movies et de documents à partir du CSV Movies small avec Python3

```python
myclient = pymongo.MongoClient("mongodb://localhost:27017/")   # connecte au MongoDB

mydb = myclient["Floupics"]   # creer la base de donnee Floupics

mycol = mydb["movies"]   # creer la collection movies
```
